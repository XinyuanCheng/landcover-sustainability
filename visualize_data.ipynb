# Visualize and understand the data

```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
DATA_DIRECTORY = '../satellite_data'
```

```python
label_names = os.listdir(DATA_DIRECTORY)
print('Labels: ')
print(label_names)
```

```python
# Get number of images per label
counter = 0
for label in label_names:
    data = os.listdir(DATA_DIRECTORY + '/' + label)
    number_files = len(data)
    counter += number_files
    print(str(label) + ': ' + str(number_files))
print()
print('Total number of images: ' + str(counter))
```

```python
[DATA_DIRECTORY + '/' + label + '/' for label in label_names]
```

```python
nrows = 4
ncols = 4
picture_index = 0
fig = plt.gcf()
fig.set_size_inches(ncols * 4, nrows * 4)

# VISUALIZE DATA FROM EACH LABEL

label_dirs = [DATA_DIRECTORY + '/' + label + '/' for label in label_names]


for label_dir in label_dirs:
    picture_index = 8
    next_picture = [os.path.join(label_dir, fname) 
                    for fname in os.listdir(label_dir)[picture_index-8:picture_index]]

    for i, img_path in enumerate(next_picture):
      sp = plt.subplot(nrows, ncols, i + 1)
      sp.axis('Off')
      img = mpimg.imread(img_path)
      plt.imshow(img)
    plt.show()
```
